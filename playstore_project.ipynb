```python
# add a column based on the value in 'Rating'          
df['High_Rating'] = np.where(df['Rating'] >= 4.5, 1, 0)
# df = df.drop(columns=['High_Rating'])
```

```python
df.head(25)
```

```python
sns.countplot(x="Category", data=df)
plt.xticks(rotation=90)
plt.title("Frequency of App Categories")
plt.show()
```

## Plotting

### 1- Histograms

```python
plt.hist(df["Rating"])
plt.xlabel("Rating")
plt.ylabel("Number of Apps")
plt.title("Distribution of App Ratings")
plt.show()
```

```python
df.hist(column='Rating', bins=100)
```

### 2- Bar Plots

```python
df['Content Rating'].unique()
```

```python
Content_Rating_count_table = pd.crosstab(index=df['Content Rating'], columns='count')
Content_Rating_count_table
```

```python
Content_Rating_count_table.plot(kind='bar', color='cyan')
plt.xticks(rotation=30)
plt.title("Content Rating")
```

### Box plot

```python
# Example: Box plot with labels and formatting
sns.boxplot(x="Category", y="Rating", data=df)
plt.xticks(rotation=90)
plt.title("Distribution of Ratings Across App Categories")
plt.show()
```

### Catplot
